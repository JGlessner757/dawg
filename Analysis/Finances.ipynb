```python
# We define the years that are included in our original dataset to do the merge.
some_values = 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020
df2 = df2.loc[df2['Year'].isin(some_values)]
```

```python
#df2 = pd.read_excel('C:/Users/jgles/dawg/Analysis/Budget_Data.xls')

def format(x):
        return '${:,.0f}'.format(x)
    
df2 = df2.rename(columns={"Column1":"Team", "Data":"University", "Year":"Season"})
df2 = df2.drop(['IPEDS ID', 'NCAA Subdivision'], axis = 1)
    
   
df2['Total Revenues'] = df2['Total Revenues'].apply(format)
df2['Total Football Spending'] = df2['Total Football Spending'].apply(format)
df2['Total Football Coaching Salaries'] = df2['Total Football Coaching Salaries'].apply(format)

df2.head()
```

```python
# Now we need to join this df to the previous one on both team and season.

df3 = df.merge(df2, on=['Team', 'Season'], how = 'outer')
df3 = df3.sort_values(by=['Total Football Spending', 'Total Football Coaching Salaries'], ascending=False, na_position='first')
df3
```
